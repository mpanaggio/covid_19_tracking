```python
from covid_plotting_tools_state import *
select,states=create_selection_box()
```

After making your selection above, select the cell below and hit "Run" or press "Shift"+ "Enter" to update the plot.

```python
%matplotlib notebook
from ipywidgets import interact, interactive, fixed, interact_manual
last_month,last_day,maxval,state,df_dict=compute_data_props(states,select)
#print(last_month,last_day,maxval)
out,ui,lines,lstart,lend=make_interactive_plot(end_month_plot=last_month,end_day_plot=last_day,
                                               def_end_month_fit=last_month,def_end_day_fit=last_day,
                                               def_month_pred=last_month,def_day_pred=last_day,
                                              maxval=maxval,country=state,df_dict=df_dict)
display(ui,out)
plt.show()
```

Note: An exponential model may not be appropriate for describing the spread of COVID-19 in every country. The data points will follow a linear trend when the y-axis is logarithmic if an exponential model is appropriate.  If the datapoints fall below the fit line on the right side of the plot, then that suggests that the growth is slower than exponential.
