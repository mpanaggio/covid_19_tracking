```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('states-daily_data.csv',parse_dates=True).rename(columns={'state':'Province/State'})\
     .drop(columns=['negative','pending','hospitalized','death','total','hash','dateChecked',
                    'totalTestResults','fips', 'deathIncrease','hospitalizedIncrease','negativeIncrease',
                    'positiveIncrease','totalTestResultsIncrease'])
cur_cols=list(df.columns).copy()
for col in cur_cols:
    if col not in ['date','Province/State','positive']:
        df=df.drop(columns=[col])
def fix_date(s):
    s=str(s)
    month=int(s[4:6])
    day=s[6:]
    year=20
    return '{}/{}/{}'.format(month,day,year)
df['date']=df['date'].apply(fix_date)
counter=0
for s in df['Province/State'].unique():
    tmp=df[df['Province/State']==s].drop(columns=['Province/State'])
    tmp=tmp.set_index('date')
    tmp=tmp.T
    tmp.index=[counter]
    tmp['Country/Region']='US'
    tmp['Province/State']=s
    tmp['Lat']=np.nan
    tmp['Long']=np.nan
    if counter==0:
        df2=tmp.copy()
    else:
        df2=pd.concat([df2,tmp],axis=0,sort=True)
    counter+=1
```

```python
tmp=df2.columns
for col in ['Country/Region', 'Province/State','Lat', 'Long']:
    tmp=np.delete(tmp, np.where(tmp == col))
print(tmp)
df2=df2[['Country/Region', 'Province/State','Lat', 'Long']+np.sort(tmp[::-1]).tolist()]
display(df2.head())
```

```python
    

df2.to_csv('state_by_state_data.csv',index=False)
```
